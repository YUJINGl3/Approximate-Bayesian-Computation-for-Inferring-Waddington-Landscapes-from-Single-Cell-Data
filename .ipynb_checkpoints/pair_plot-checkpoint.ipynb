```python
# this file output the figure of pair plot of joint posterior distribution for different distance metrics
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#load joint posterior distribution
mmd_dens = pd.read_csv('mmd.csv', header = None)
bhatta_dens = pd.read_csv('bhatta.csv', header = None)
opt_dens = pd.read_csv('optimal_transport.csv', header = None)
```

```python
sns.set(font_scale=2)
g = sns.pairplot(opt_dens, corner = True, kind="hist")
g.set(xticklabels=[])
g.set(yticklabels=[])
plt.savefig('./Figure/output_opt.png')
```

```python
sns.set(font_scale=2)
g = sns.pairplot(mmd_dens, corner = True, kind="hist")
g.set(xticklabels=[])
g.set(yticklabels=[])
plt.savefig('./Figure/output_mmd.png')
```

```python
sns.set(font_scale=2)
g = sns.pairplot(bhatta_dens, corner = True, kind="hist")
g.set(xticklabels=[])
g.set(yticklabels=[])
plt.savefig('./Figure/output_bhatta.png')
```
